```python
# using the search endpoint, let's get information about every star trek show
# http://api.tvmaze.com/search/shows?q=<querystring>
# ? signals that you are passing arguments to your endpoint
# each argument has a variable name followed by = sign followed by the value of that argument
starTrekQuery = requests.get(rootURL+"/search/shows?q=Star Trek").json()
```

```python
starTrekQuery
```

```python
# how many results came back?
len(starTrekQuery)
```

```python
# let's iterate through the results and get information about each show
for showDetails in starTrekQuery:
    # showDetails is what kind of datatype? a dictionary that contains information about each show
    print(showDetails["show"]["name"])
    # print the genres in a comma separated list
    print(", ".join(showDetails["show"]["genres"]))
    print(showDetails["show"]["summary"])
    print()
```

```python
# let's use the information returned fro the first query from each show to get more information about that show
# episodes endpoint
# http://api.tvmaze.com/shows/<id>/episodes
# the id is a *unique identifier*
```

```python
for showDetails in starTrekQuery:
    print(showDetails["show"]["name"])
    # call the episodes endpoint using the id of the current show
    print(showDetails["show"]["id"])
    #print(rootURL+"/shows/"+str(showDetails["show"]["id"])+"/episodes")
    episodeQuery = requests.get(rootURL+"shows/"+str(showDetails["show"]["id"])+"/episodes").json()
    #print(episodeQuery)
    # episodeQuery gives us a list of dictionaries
    # let's iterate through those dictionaries
    for episode in episodeQuery:
        # episode is going to store a dictionary that contains information about each episode of the show
        # one at a time
        print("Episode Number: "+str(episode["number"]))
        print("Season Number:" +str(episode["season"]))
        print("Episode Summary: "+str(episode["summary"])) # cast summary as a str to handle the None
        print()
    print()
```

```python
# let's write a function that can grab the top N albums for a given artist
def printTopNAlbums(artistName,numAlbums):
    artistTopAlbumQuery = requests.get(rootURL+
        "?method=artist.gettopalbums&api_key="+aKey+"&format=json&artist="+artistName+
        "&limit="+str(numAlbums)).json()
    for album in artistTopAlbumQuery["topalbums"]["album"]:
        print(album["name"])
```

```python
import urllib
def getABunchOfStuffAboutAnArtist(artistName,numAlbums):
    
    # adding error checking
    # check numAlbums to make sure it is an int
    if (type(numAlbums) != int):
        print("The number of albums must be an integer")
        
    else:
        # create intial query for artistName
        artistQuery = requests.get(rootURL+
            "?method=artist.getinfo&api_key="+aKey+"&format=json&artist="+urllib.parse.quote(artistName)).json()

        # check to make sure we got artist data back from the query
        if ("error" in artistQuery):
            #print("Artist "+artistName+ " does not appear to be in the dictionary")
            print(artistQuery["message"])
        else:
            # display information about the artist
            print("Artist Name: "+artistName)
            print("Artist Summary: "+artistQuery["artist"]["bio"]["summary"])

            # grab top albums for the artist
            print()
            print("Top "+str(numAlbums)+ " for "+artistName+":")
            printTopNAlbums(urllib.parse.quote(artistName),numAlbums)

            # let's look at similar artists
            print()
            print("***")
            print()
            print("Here are the artists similar to "+artistName+
                  " and their top "+str(numAlbums)+" albums:")
            similarArtists = artistQuery["artist"]["similar"]["artist"] # list of similar artists

            # check if similarArtists is an empty list
            if (similarArtists == []):
                print("There's no one like "+artistName)
            else:
                # iterate through list of similar artists
                for simArtist in similarArtists:
                    print("Artist Name: "+simArtist["name"])
                    print()
                    printTopNAlbums(urllib.parse.quote(simArtist["name"]),numAlbums)
                    print()
```
